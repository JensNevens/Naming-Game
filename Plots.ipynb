```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')

fig = plt.figure()
plt.title('The Naming Game')
plt.ylim((0,1.1))

ax1 = df.R.plot(style='b', label='Success Rate')
ax1.set_ylabel('Success Rate / Coherence')

ax2 = df.C.plot(style='r', label='Coherence')

ax3 = df.S.plot(secondary_y=True, style='g', label='Size')
ax3.set_ylabel('Lexicon Size')

handles,labels = [],[]
for ax in fig.axes:
    handle,label = ax.get_legend_handles_labels()
    for h,l in zip(handle,label):
        handles.append(h)
        labels.append(l)

plt.legend(handles,labels,loc=4,fancybox=True,shadow=True)

#plt.savefig('./plots/Game2500_Agent10_Meaning10.pdf')
```
